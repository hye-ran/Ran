####_**인공지능의 정의**_
#####인공지능 (AI) 란 인간의 지능으로 할 수 있는 사고, 학습, 자기계발 등을 컴퓨터가 할 수 있도록 하는 방법을 연구하는 컴퓨터 공학 및 정보기술의 한 분야로서, 컴퓨터가 인간의 지능적인 행동을 모방할 수 있도록 하는 것을 인공지능이라고 할 수 있다

#1.언어
* 구글 번역기(Google Translate)
    
    Google의 무료 서비스로 영어와 100개 이상의 다른 언어간에 단어, 구문, 웹페이지를 즉시 번역 한다.
    

> 2007년 구글에서 발표한 무료 번역(텍스트 번역) 서비스로, 현재 세계에서 가장 많이 사용되는 기계번역 서비스 이다. 2016년 9월 28일 부터 알파고에도 쓰인 딥러닝 기술을 적용한다. 기존 구글 번역의 오류를 58~87% 줄였다고 한다.2016년 11월 15일 인공신경망 기반 영어↔한국어 번역 서비스가 도입되었으며(신경 기계 언어 엔진인 구글 신경 기계 번역(GNMT)으로 전환) 번역품질이 크게 향상되었다 당장 차이를 매우 확연히 느낄 수 있을 정도이며 상당 수 문장을 완벽에 가깝게 번역한다 

이처럼 인공지능 번역기는 문장을 이해하고 분석하는 능력이 웬만한 사람못지   않아 언어의 장벽을 허물고 있다. 앞으로 더 진화될 인공지능 번역기가 기대된다.
그밖에 흥미로웠던 인공지능 언어 분야의 사례로는 대구경북첨단의료재단의 창업 지원을 받아 언어장애를 가진 환자들이 가정에서 피드백을 받으며 언어장애 치료를 돕는 음성 인식 인공지능이 탑재된 '토키토키' 앱이다. 인공지능이 소외된계층에게도 도움이 많이 될수 있다는 점이 무척 인상 깊었다.

#4.자율주행 

* 미국 자동차 회사 로컬 모터스-올리(Olli) 미니 버스



> 이 버스는 스마트폰으로 출발지와 목적지를 입력하면 내가 있는 곳으로 와 나를 태우고 목적지까지 데려다 준다. 올리는 IBM의 슈퍼컴퓨터 플랫폼 왓슨을 내장해 탑승자와 대화도 가능하다. 최대 12명의 승객이 탈 수 있다. 


한국도 자율주행 버스에 대한 실험이 시작됐다. 2018년 경기도가 제작한 운전자 없는 자율주행차 '제로 셔틀'이 일반도로를 달렸다. 제로셔틀은 경기도가 차세대융합기술연구원에 의뢰해 3년간 연구 끝에 개발한 자율주행차다.
11인승 미니버스로 2020년 말까지 판교 제2테크노밸리 기업지원허브 입구에서 판교역까지 5.5km 구간(30분 소요)을 시속 25km 이내로 순환하며 시범 운행하게 된다.
제로셔틀에는 운전대, 액셀, 브레이크 페달이 없다. 통합관제센터와 교통신호 정보, GPS 위치 보정 정보신호, 주행안전정보등을 무선으로 주고받는 차량사물통신기술을 활용해 운행한다. 

   이번에 인공지능 기술이 사용되는 제품과 서비스를 알아보면서 인공지능이 정말 우리생활과 밀접하게 다양하게 활용된다는것을 알 수 가 있었다. 이세돌과 알파고의 대국을 보면서 인공지능이 인간을 위협할 수 있겠다는 생각을 한 적도 있는데 인공지능에 대해 공부하면서 인간에게 주는 긍정적인 측면이 더 많다는것을 알게되었고 너무 두려워할 필요가 없다는것도 알게되었다. 앞으로 인공지능의 발전이 기대가 된다.
